```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('coin_Cardano.csv')
df
```

```python
df=df.set_index(pd.DatetimeIndex(df['Date'].values))
df
```

```python

df=df[['Close']]
df.head(10)
```

```python
pred_day=7
df['Prediction']=df[['Close']].shift(-pred_day)
df
```

```python
df = df[df['Prediction'].notnull()]
df
```

```python
X=np.array(df.iloc[:,:1])
X
```

```python
y=np.array(df['Prediction'])
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(max_depth=15, random_state=500)
regr.fit(X, y)
regr.score(X_test,y_test)
```

```python
pred=regr.predict([X_test[10]])
print("Prediction price")
print(pred)
print("Real price")
print([y_test[10]])
```
